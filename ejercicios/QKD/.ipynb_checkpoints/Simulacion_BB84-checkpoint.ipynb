```python

import pandas as pd
import random

# Función para elegir aleatoriamente una base
def random_base():
    return random.choice(["↕", "↗"])

# Simulación de n rondas de transmisión usando BB84
def simulate_bb84_rounds(n=15):
    data = []
    for i in range(1, n + 1):
        bit_alice = random.randint(0, 1)
        base_alice = random_base()
        photon = bit_alice  # simplificación

        base_bob = random_base()
        if base_alice == base_bob:
            bit_bob = photon
            match = "sí"
            use = "sí"
        else:
            bit_bob = random.randint(0, 1)
            match = "no"
            use = "no"

        data.append({
            "N°": i,
            "Bit de Alice": bit_alice,
            "Base de Alice (↕/↗)": base_alice,
            "Fotón enviado": photon,
            "Base de Bob (↕/↗)": base_bob,
            "Bit recibido": bit_bob,
            "¿Bases coinciden?": match,
            "¿Usar bit?": use
        })

    df = pd.DataFrame(data)
    usable_bits = len(df[df["¿Usar bit?"] == "sí"])
    total_bits = len(df)
    percentage = round((usable_bits / total_bits) * 100, 2)

    print(f"Bits usados para la clave: {usable_bits} / {total_bits} ({percentage}%)")
    return df

# Ejecutar simulación
df_bb84 = simulate_bb84_rounds()
df_bb84
```
